### pandas datafrmes
- convertion of table format
- creating dataframes from dictionary
- creating dataframes from arrays

```python
#### from dictionary
studentmarks={"names":['ram','asha','giri'],
              "maths":[23,45,67],
              "sci":[56,36,98],
              "eng":[23,56,78]}
studentmarks
```

```python
pd.DataFrame(studentmarks)
```

```python
### from arrays
a1=np.array([['ram',23,45,67],['asha',56,36,98]])
pd.DataFrame(a1,columns=['name','maths','sci','eng'])
```

```python
a1=np.array([['ram',23,45,67],['asha',56,36,98]])
pd.DataFrame(a1,columns=['name','maths','sci','eng'],index=['1','2'])
```
